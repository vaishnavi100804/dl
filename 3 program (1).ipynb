```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("iris.csv")
```

```python
df
```

```python
x=df.iloc[:, 0:2].values
y=df.iloc[:, -1].values
```

```python
x
```

```python
y
```

```python
y_binary=np.where(y=='iris_setosa', 1.0, 0.0)
```

```python
y_binary
```

```python
X=(x-x.mean(axis=0))/x.std(axis=0)
```

```python
X
```

```python
x_tf=tf.constant(X, dtype='float32')
```

```python
x_tf
```

```python
w=tf.constant([[1.0],
               [-1.0]], dtype='float32')
b=tf.constant([0.2], dtype='float32')
```

```python
w
```

```python
b
```

```python
z=tf.matmul(x_tf, w)+b
```

```python
z
```

```python
y_pred_step=(z>=0, 'int 32')
```

```python
y_pred_sigmoid=tf.math.sigmoid(z)
```

```python
print("first 10 True labels (Setosa=1, others=0):", y_binary[:10])
print("first 10 step outputs:", y_pred_step [:10])
print("first 10 sigmoid probabilities:", y_pred_sigmoid [:10])
```
